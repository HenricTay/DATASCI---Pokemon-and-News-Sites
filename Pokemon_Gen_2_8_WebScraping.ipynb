# Getting data directly from Bulbapedia using `requests` and `BeautifulSoup`


```python
import requests

URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
```

```python
page=requests.get(URL)
```

```python
print(page.content)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Get main content <div>
poke_content=soup.find(id='mw-content-text')

# Get all <table> elements
poke_tables=poke_content.find_all('table')
```

### Get list of all Generation Pokemons

```python
gen_list=poke_tables[1]
```

```python
# Check its contents and find where the first Pokemon entry is
gen_list.contents
```

```python
# The first Pokemon entry
gen_list.contents[3]
```

```python
info_start=3

# Let's figure out how to get each item for Bulbasaur
info_row=gen_list.contents[info_start]

for i in range(len(info_row.contents)):
    print(f'Index {i} - {info_row.contents[i]}')
```

```python
info_row.contents[7].a.get('href')
```

```python
# Extract items of interest

link = 'https://bulbapedia.bulbagarden.net' + info_row.contents[7].a.get('href')
kdex=info_row.contents[1].text.strip()
ndex=info_row.contents[3].text.strip()
name=info_row.contents[7].text.strip()
type1=info_row.contents[9].text.strip()

print(f'Pokemon {ndex} {name} is a {type1} Pokemon. Link: {link}')
```

### Get all Pokemons from Gen 2 to Gen 8


```python
for x in range(2,9):
    gen_list=poke_tables[x]
    for i in range(info_start, len(gen_list.contents), 2):
      poke_info=gen_list.contents[i]
      kdex=poke_info.contents[1].text.strip()
      ndex=poke_info.contents[3].text.strip()
      name=poke_info.contents[7].text.strip()
      type1=poke_info.contents[9].text.strip()
      link = 'https://bulbapedia.bulbagarden.net' + poke_info.contents[7].a.get('href')
      if len(poke_info.contents) > 10:
          type2=poke_info.contents[11].text.strip()
          print(f'Pokemon {ndex} {name} is a {type1} & {type2} Pokemon. Link: {link}')
      else:
          print(f'Pokemon {ndex} {name} is a {type1} Pokemon. Link: {link}')
```

### Save them in a JSON

```python
gen_json = []
for x in range(2,9):
    gen_list=poke_tables[x]
    for i in range(info_start, len(gen_list.contents), 2):
        poke_info=gen_list.contents[i]
        kdex=poke_info.contents[1].text.strip()
        ndex=poke_info.contents[3].text.strip()
        name=poke_info.contents[7].text.strip()
        type1=poke_info.contents[9].text.strip()
        link = 'https://bulbapedia.bulbagarden.net' + poke_info.contents[7].a.get('href')
        if len(poke_info.contents) > 10:
            type2=poke_info.contents[11].text.strip()
            if x == 2:
              gen_json.append({
                  "Jdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "type2": type2,
                  "link" : link,
              })
            elif x == 3:
              gen_json.append({
                  "Hdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "type2": type2,
                  "link" : link,
              })
            elif x == 4:
              gen_json.append({
                  "Sdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "type2": type2,
                  "link" : link,
              })
            elif x == 5:
              gen_json.append({
                  "Udex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "type2": type2,
                  "link" : link,
              })
            elif x == 6:
              gen_json.append({
                  "Ce/Co/Mo Kdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "type2": type2,
                  "link" : link,
              })
            elif x == 7:
              gen_json.append({
                  "Adex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "type2": type2,
                  "link" : link,
              })
            elif x == 8:
              gen_json.append({
                  "Gdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "type2": type2,
                  "link" : link,
              })
        else:
            if x == 2:
              gen_json.append({
                  "Jdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "link" : link,
              })
            elif x == 3:
              gen_json.append({
                  "Hdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "link" : link,
              })
            elif x == 4:
              gen_json.append({
                  "Sdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "link" : link,
              })
            elif x == 5:
              gen_json.append({
                  "Udex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "link" : link,
              })
            elif x == 6:
              gen_json.append({
                  "Ce/Co/Mo Kdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "link" : link,
              })
            elif x == 7:
              gen_json.append({
                  "Adex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "link" : link,
              })
            elif x == 8:
              gen_json.append({
                  "Gdex": kdex,
                  "Ndex": ndex,
                  "name": name,
                  "type1": type1,
                  "link" : link,
              })
        
gen_json
```

```python
import json

with open('Pokemon Generations 2-8.json', 'w', encoding='utf-8') as f:
    json.dump(gen_json, f, ensure_ascii=False, indent=4)
```
