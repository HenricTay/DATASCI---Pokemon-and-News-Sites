#Getting the total number of articles from newsapi.get_everything



```
news_list = articles['articles']

news_list
```

```
news_json = []

for news in news_list:
      #  date = news['publishedAt']
      #  title= news['title']
      #  full_article= news['content']
      #  author= news['author']
        news_json.append({
            "Date": news['publishedAt'],
            "Title": news['title'],
            "Full_Article":news['content'],
            "Author":news['author'],
            "From": news['source']['name']
        })

len(news_json)
```

```
news_json
```

```
import json
with open('TheVerge articles 03-11-21 to 03-12-21.json', 'w', encoding='utf-8') as f:
    json.dump(news_json, f, ensure_ascii=False, indent=4)
```
